```python
import chess
import chess.pgn
import io
```

```python
def pgn_to_fen(pgn_string):
    game = chess.pgn.read_game(io.StringIO(pgn_string))

    fen_positions = []
    board = game.board()
    for move in game.mainline_moves():
        fen_positions.append(board.fen())
        board.push(move)

    return fen_positions
        
```
